# King County Washington Housing Analysis

#### Overview:
This project is a multiple regression modeling project using the King County Washington Housing Dataset.  The end goal of this project is to discern what aspects of their properties homeowners may improve to increase their property value.

#### Business Understanding:
This analysis is for Realty, Inc.  Realty is looking for ways to entice homeowners to list their properties for sale on the market.  To do so, they need strong recommendations to offer homeowners.  This analysis delves into the King County Housing Dataset to provide solutions for Realty. 

#### Modeling:
The model is a multiple linear regression built in statsmodel.  Price is considered as the dependent variable.  The end result consisted of square footage, grade, and bedrooms.  The model satisfies all multiple regression assumtions, P- values are below 0.05, so the null hypothesis was rejected.  


#### Conclusion
The model shows that homeowners should increase their square footage, add bedrooms, and both improve the condition of their houses.  The model does have some limitations.  First, the data was logarithmically transformed, so any new data would have to be treated in the same manner.  The model does not take into account any other area except King County. New models would have to be created for data outside of King County.

Further research should be done regarding the optimal time of the year to sell, as well as the effects of crime throughout the county.
